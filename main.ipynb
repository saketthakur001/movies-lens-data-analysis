```python
import imdb

# create an instance of the IMDb class
ia = imdb.IMDb()

# get the movie title and poster URL
def get_movie_title_and_poster(movie_id):
    movie = ia.get_movie(movie_id)
    # mediam size cover url
    # return movie['title'], movie['cover url']
    return movie['title'], movie['full-size cover url']
```

```python
import pandas as pd
new_df = pd.read_csv("new_df.csv")

# all the that has more than 150 count
new_df = new_df[new_df['count'] > 150]

# sort by highest rating and lowest count
new_df = new_df.sort_values(by=['rating'], ascending=[False])

# top 1000 movies
top_movies = new_df.head(1000)
# iterate though each top_movies and get the title and poster URL
for index, row in top_movies.iterrows():
    try:
        title, poster = get_movie_title_and_poster(row['imdbId'])
        top_movies.loc[index, 'title'] = title
        top_movies.loc[index, 'poster'] = poster
    except:
        print('Error in getting title and poster for movie id: ', row['imdbId'])
```
